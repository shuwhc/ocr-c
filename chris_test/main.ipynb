```python
from PIL import Image
import pytesseract
from pytesseract import Output
import cv2
import numpy as np
from matplotlib import pyplot as plt
import re
```

```python
image = cv2.imread("D:/School/School_sem3/IA/IA_project/test_3.jpg") #, cv2.IMREAD_GRAYSCALE
#img_RGB = cv2.cvtColor(image, cv2.COLOR_BAYER_BG2BGR)
cv2.resize(image ,(600,800),interpolation= cv2.INTER_CUBIC)
Simple_Thresholding = cv2.threshold(image, 110, 255, cv2.THRESH_BINARY_INV)[1]

results = pytesseract.image_to_boxes(image)
ih, iw, ic = image.shape
for box in results.splitlines():
    box = box.split(" ")
    print(box)

    x, y, w, h = int(box[1]), int(box[2]), int(box[3]), int(box[4])
    cv2.rectangle(image, (x,ih-y), (w,ih-h), (0,255,0), 2 )
    cv2.putText(image, box[0],(x, ih-h), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 1)
```

```python
cv2.imshow("input", image)
cv2.waitKey(0)
```
